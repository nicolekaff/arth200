```python
import pandas as pd
import sys
```

```python
df = pd.read_excel("images.xlsx")
df= df.rename(columns=str.lower)
df
```

```python
# Process architecture
for i in df.index:
    name = df.iloc[i]['image']
    prefix = name.split(".")[0]
    if df.iloc[i]['category'] == 'Architecture':
        df.loc[i].to_json("{}.json".format(prefix))
```

```python
# Process non-architecture
for i in df.index:
    name = df.iloc[i]['image']
    prefix = name.split(".")[0]
    if df.iloc[i]['category'] != 'Architecture':
        df.loc[i].to_json("{}.json".format(prefix))
```
